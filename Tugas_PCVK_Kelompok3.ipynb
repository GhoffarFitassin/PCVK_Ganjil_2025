```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

```python
# Persiapan kartu
img_path = "KTM_Card.jpeg"
img = cv.imread(img_path)

# Check if the image was loaded successfully
if img is None:
    print(f"Error: Could not load image from {img_path}")
else:
    print(f"Image loaded successfully from {img_path}")
```

```python
resized_img = cv.resize(img, (500, 300))
cv2_imshow(resized_img)
```

```python
# Convert BGR (OpenCV) to RGB (matplotlib)
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
# Tentukan koordinat area teks "POLITEKNIK"
xP, yP, wP, hP = 150, 38, 200, 40
# Tentukan koordinat area teks "NEGERI"
xN, yN, wN, hN = 150, 83, 95, 40
# Tentukan koordinat area teks "MALANG"
xM, yM, wM, hM = 255, 83, 120, 40
```

```python
# Tutup area (tulisan POLITEKNIK) dengan warna kuning
img_rgb[yP:yP+hP, xP:xP+wP] = (255, 255, 0)
# Tutup area (tulisan NEGERI) dengan warna merah
img_rgb[yN:yN+hN, xN:xN+wN] = (255, 0, 0)
# Tutup area (tulisan MALANG) dengan warna hitam
img_rgb[yM:yM+hM, xM:xM+wM] = (0, 0, 0)

# Tampilkan gambar
plt.imshow(img_rgb)
plt.axis("off")
plt.show()
```
